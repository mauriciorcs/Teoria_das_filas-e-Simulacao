```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
resultados_simulador = pd.read_excel("q2_resultados.xlsx", sheet_name="resultado_simulador")
resultados_modelo = pd.read_excel("q2_resultados.xlsx", sheet_name="resultado_modelo")
```

```python
resultados_simulador
```

```python
resultados_modelo
```

## Questão 2
### Gráfico de linhas comparando o tempo médio de resposta


```python
# Criar um DataFrame combinado com uma coluna indicando a origem dos dados
resultados_simulador['Planilha'] = 'resultado_simulador'
resultados_modelo['Planilha'] = 'resultado_modelo'
dados_combinados = pd.concat([resultados_simulador, resultados_modelo])
```

```python
sns.lineplot(data=dados_combinados, x='TaxaDeChegadaMedia', y='TempoMedioDeResposta', hue= 'Planilha', marker='o')

# Adicionar rótulos e título
plt.xlabel('Taxa Média de Chegada')
plt.ylabel('Tempo Médio de Resposta')
plt.title('Comparação do Tempo Médio de Resposta')
plt.legend(title='', loc='best')
plt.savefig('q2_tempo_medio_resposta.png')

plt.show()
```

### Tamanho Medio da Fila

```python
sns.lineplot(data=dados_combinados, x='TaxaDeChegadaMedia', y='TamanhoMedioDaFila', hue= 'Planilha', marker='o')

plt.xlabel('Taxa Média de Chegada')
plt.ylabel('Tamanho da Fila')
plt.title('Comparação do Tamanho Médio da Fila')
plt.legend(title='', loc='best')
# plt.savefig('tamanho_medio_fila.png')

plt.show()
```

## Questão 3

```python
resultados_simulador_2 = pd.read_excel("q3_resultados.xlsx", sheet_name="resultado_simulador")
resultados_modelo_2 = pd.read_excel("q3_resultados.xlsx", sheet_name="resultado_modelo")
```

```python
# Criar um DataFrame combinado com uma coluna indicando a origem dos dados
resultados_simulador_2['Planilha'] = 'resultado_simulador'
resultados_modelo_2['Planilha'] = 'resultado_modelo'
dados_combinados_2 = pd.concat([resultados_simulador_2, resultados_modelo_2])
dados_combinados.sample(3)
```

```python
sns.lineplot(data=dados_combinados_2, x='NumServidores', y='TempoMedioDeResposta', hue='Planilha', marker='o')

meta_tempo_resposta = 1  
plt.axhline(y=meta_tempo_resposta, color='r', linestyle='--', label='Meta de Tempo Médio de Resposta')

# Adicionar rótulos e título
plt.xlabel('Número de Servidores')
plt.ylabel('Tempo Médio de Resposta')
plt.title('Comparação do Tempo Médio de Resposta')

plt.legend(title='', loc='best')
plt.xticks([50, 30, 20, 15, 10, 7], rotation=45)

plt.gca().invert_xaxis()

plt.savefig('q3_tempo_medio_resposta.png')

plt.show()
```

```python
sns.lineplot(data=dados_combinados_2, x='NumServidores', y='TamanhoMedioDaFila', hue= 'Planilha', marker='o')

plt.xlabel('Número de Servidores')
plt.ylabel('Tamanho da Fila')
plt.title('Comparação Tamanho Médio da Fila')
plt.legend(title='', loc='best')
plt.gca().invert_xaxis()

plt.show()
```
